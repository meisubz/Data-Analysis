```python
#Importing the dependencies
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

### Data Collection & Analysis

```python
#Importing the csv file
sonar_ds = pd.read_csv('sonar data.csv',header=None)
```

```python
#Displaying first 5 rows of data
sonar_ds.head()
```

```python
#Checking number of rows & columns
sonar_ds.shape
```

```python
#Displaying last 5 rows of data
sonar_ds.tail()
```

```python
#The distribution of rocks & mines in given dataset
x = sonar_ds[60]
x.value_counts()
```

```python
#Based on the last column use groupby for the rest of the columns and find mean value
#sonar_ds.groupby(60).mean()
sonar_ds.groupby(by=sonar_ds[60]).mean()
```

```python
#Separating input feature & target variable/value
X = sonar_ds.drop(columns=60,axis=1)
Y = sonar_ds[60]
```

```python
#Printing the Input feature
print(X)
```

```python
#Printing the Target Variable
print(Y)
```

```python
#Splitting data into testing & training data
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, Y_test.shape)
```

```python
#Model Training
model = LogisticRegression()
#Training the ML Model using training data
model.fit(X_train, Y_train)
```

```python
#Model Evaluation

#Check Accuracy Score of training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy Score of training data:",training_data_accuracy)
```

```python
#Check Accuracy Score of testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy Score of testing data:",testing_data_accuracy)
```
