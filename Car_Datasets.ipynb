```python
import pandas as pd
df = pd.read_csv('USA_cars_datasets.csv')
df
```

```python
#Check basic info of the dataset
df.info()
```

```python
#Removing the undesired column
df.drop('Unnamed: 13',axis=1,inplace=True)
```

```python
#Find unique states from dataset and its count
c = pd.unique(df['state'])
print(c,'\n')
print(len(c))
```

```python
#Find the unique brands
z = pd.unique(df['brand'])
len(z)
print('\n',z)
```

```python
#Filter data: Count of all rows having state New Jersey
df[df['state']=='new jersey']['state'].count()
```

```python
#Find average price of all brands
df.groupby('brand')['price'].mean().sort_values()
```

```python
#Create a new column with vehicle age
from datetime import datetime
current_year = datetime.now().year
df['Vehicle Age'] = (current_year-df['year'])
df
```

```python
#Find all cars manufactured in 2008
df[df['year'] == 2008]['year'].count()
```

```python
#Find all the cars manufactured in 2008 & color of car should be either red, blue or yellow
df[(df['year']==2008) & (df['color'].isin(['red','blue','yellow']))]
```

```python
#Find count of different colors car in the state New Jersey
df[df['state']=='new jersey']['color'].value_counts()
```

```python
#Find which state has the highest yellow cars
#df[df['color']=='yellow']['state'].value_counts()
yellow_cars = df[df['color']=='yellow']
yellow_cars.groupby(by='state')['state'].count()
```

```python
#Find which state has highest cars where brand is toyota or ford or nissan & color should be either black or silver
df[(df['brand'].isin(['toyota', 'ford', 'nissan'])) & (df['color'].isin(['black', 'silver']))]['state'].value_counts()
```

```python
#Find the count of all colors in all states and sort them according to the highest number to lowest.
df.groupby(by='state')['color'].value_counts().sort_values(ascending=False)
```

```python
#Filter the entire data where price is greater than $25000
df[df['price']>25000]
```

```python
#Find unique values of title_status column and count the number of clean vehicle
print(pd.unique(df['title_status']))
print()
df[df['title_status']=='clean vehicle']['title_status'].count()
```

```python
import matplotlib.pyplot as plt
#Pie Chart
x = ['clean vehicle', 'salvage insurance']
plt.title('The Status Ratio Pie Chart')
y = df.groupby(by='title_status')['title_status'].count()
plt.style.use('dark_background')
plt.pie(y, labels=x,autopct='%1.1f%%',colors=['skyblue','maroon'])
plt.show()
```

```python
x=['white','black','silver','blue','gold','ruby red']
y=df[(df['color'].isin(['blue','gold','silver','ruby red','white','black']))]["color"].value_counts()
plt.title('The Color Count',fontsize=15)
plt.style.use('ggplot')
plt.xlabel('The Color')
plt.ylabel('The Count')
plt.title('The Color Count Bar Graph',color='gray')
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center',color='black')
plt.bar(x,y,color='gold',edgecolor='white',alpha=0.5)
addlabels(x, y)
plt.show()
```
