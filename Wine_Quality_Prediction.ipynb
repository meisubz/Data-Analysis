```python
#Importing the Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

#Importing the Visualization tools
import matplotlib.pyplot as plt
import seaborn as sns

#Model Selection
from sklearn.ensemble import RandomForestClassifier

#Importing accuracy score function
from sklearn.metrics import accuracy_score
```

### Setting Up the Data

```python
#Importing the dataset
wine_ds = pd.read_csv('wine.csv')
wine_ds
```

```python
#Print the first 5 rows
wine_ds.head()
```

```python
#Print the last 5 rows
wine_ds.tail()
```

```python
#Number of rows & columns
wine_ds.shape
```

```python
#Getting information about the data
wine_ds.info()
```

```python
#Find the missing values (if any)
wine_ds.isnull().sum()
```

```python
#Gathering Statistical measure of the data
wine_ds.describe()
```

```python
#Unique Qualities of the Wine from the dataset
print(pd.unique(wine_ds['quality']),'\n')
wine_ds['quality'].value_counts()
```

```python
sns.catplot(x='quality',data=wine_ds,kind='count')
```

### Creating the Model

```python
#Separating the feature & target columns
#Input Feature: x(All columns except quality) & Target Column: y(quality column)
X = wine_ds.drop(columns='quality',axis=1)
Y = wine_ds['quality']
```

```python
#Printing the Input feature
print(X)
```

```python
#Printing the Target Variable
print(Y)
```

```python
#Splitting data into testing & training data
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, Y_test.shape)
```

### Training the Model

```python
#Model Training
model = RandomForestClassifier()
#Training the ML Model using training data
model.fit(X_train, Y_train)
```

```python
#Model Evaluation

#Check Accuracy Score of training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy Score of training data:",training_data_accuracy)
```

```python
#Check Accuracy Score of testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy Score of testing data:",testing_data_accuracy)
```

```python
#Building a Predictive System
input_data = (7.9,0.35,0.46,3.6,0.078,15.0,37.0,0.9973,3.35,0.86,12.8)
#Converting the input data into a numpy array
input_data_as_numpy_array = np.asarray(input_data)
#Reshape the numpy array as we are predicting for one data point
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)
```

```python
#Final Display of the prediction
prediction = model.predict(input_data_reshape)
print("The quality of the red wine according to the prediction:",prediction)
```
