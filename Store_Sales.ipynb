```python
import pandas as pd
```

```python
#Reading the excel file
df = pd.read_excel('Store_Sales.xlsx')
df
```

```python
#Information about the data
df.info()
```

```python
#Find the count of orders where order priority is either High or Critical & is from Home Office Customer Segment
df[(df['Order Priority'].isin(["High","Critical"])&(df["Customer Segment"]=="Home Office"))]["Customer Segment"].count()
```

```python
#Find total sum of orders placed in the year 2015
#df['Year']=df['Order Date'].dt.year
df[df['Year']==2015].count()
```

```python
#Find sum of orders placed in the first quarter of 2015
#Year (Jan-March) (Apr-June) (July-Sep) (Oct-Dec)
#df['Month'] = df['Order Date'].dt.month
df[df['Month'].isin([1,2,3])]['Quantity ordered new'].count()
```

```python
#Find the total sum of orders placed in the month June & year 2015
df[(df['Month']==6) & (df['Year']==2015)]['Customer ID'].count()
```

```python
#Access first 50 rows & analyse the count of orders from city "Prior Lake"
cond = df.iloc[0:51]
cond[cond['City']=="Prior Lake"]
```

```python
#Access all the rows except last 20 rows & find out the count of orders where Product Container is Wrap Bag
con = df.tail(-20)
con[con['Product Container']=='Wrap Bag'].count()
```
