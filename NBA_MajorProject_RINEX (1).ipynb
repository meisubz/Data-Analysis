```python
#Data Visualization
import matplotlib.pyplot as plt
```

```python
#Pie Chart for The Salary & Position
x = ['C','PF','PG','SF','SG']
ex = [0,0,0,0,0.1]
plt.style.use('seaborn-dark-palette')
plt.title('The Salary & Position Pie Chart')
y = df.groupby(by='Position')['Salary'].count()
plt.pie(y,labels=x,autopct='%1.1f%%',explode=ex)
plt.show()
```

```python
#Using a bar graph to visualise The Position & Weight Comparison

plt.figure(figsize = (8, 3))
x = ['C','PF','PG','SF','SG']
plt.title('The Position & Weight Comparison',fontsize=15,color='gray',fontweight="bold")
y = df.groupby(by='Position')['Weight(lbs)'].count()
plt.xlabel('Positions')
plt.ylabel('Weight (in lbs)')
plt.style.use('ggplot')
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center',color='black')
plt.bar(x,y,color='maroon',edgecolor='yellow',alpha=0.8,width=0.5)
addlabels(x, y)
plt.show()
```

```python
#Using a bar graph to visualise The College Count
plt.figure(figsize = (10, 5))
y = df[(df['College'].isin(['Texas','Kansas', 'Memphis', 'LSU','UCLA']))]["College"].value_counts()
x = ['Kansas','UCLA','Texas','LSU','Memphis']
plt.title('The College Count',fontsize=15,fontweight="bold")
plt.xlabel('College Name')
plt.ylabel('The Count')
plt.style.use('dark_background')
plt.ylim(0,20)
plt.bar(x,y,color='salmon',edgecolor='white',alpha=0.8)
plt.show()
```
